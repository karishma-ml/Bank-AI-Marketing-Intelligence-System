```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv("bank_data.csv", sep = ";")
# because data is seperate by the double quotes and comma 
```

```python
df.head()
```

```python
pd.set_option("display.max_columns", None)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# converting unknown values into null values
df.replace("unknown", np.nan, inplace= True)
```

```python
df.isnull().sum()
```

```python
# replacing null values
df["job"] = df["job"].fillna(df["job"].mode()[0])
df["marital"] = df["marital"].fillna(df["marital"].mode()[0])
df["education"] = df["education"].fillna(df["education"].mode()[0])
df["default"] = df["default"].fillna(df["default"].mode()[0])
df["housing"] = df["housing"].fillna(df["housing"].mode()[0])
df["loan"] = df["loan"].fillna(df["loan"].mode()[0])
```

```python
df.isnull().sum()
```

```python
from catboost import CatBoostClassifier
```

```python
from catboost import Pool
```

```python
x = df.drop("y", axis = 1)
y = df["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify= y)
```

```python
len(x_train), len(x_test)
```

```python
df[:1]
```

```python
pool_train = Pool(x_train, y_train, cat_features=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome'])
```

```python
pool_test = Pool(x_test, y_test, cat_features=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome'])
```

```python
cbc = CatBoostClassifier(iterations=100)
```

```python
cbc.fit(pool_train)
```

```python
y_pred = cbc.predict(pool_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
import joblib
```

```python
joblib.dump(cbc, "catboost.pkl")
```
